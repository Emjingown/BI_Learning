```python
import numpy as np
from scipy.linalg import svd
from PIL import Image
import matplotlib.pyplot as plt
```

## 图像灰度化

### 加载和显示原图像

```python
# 加载256色图片
image = Image.open('./华工之夜.jpg') 
A = np.array(image)
# 显示原图像
plt.imshow(A, cmap=plt.cm.gray, interpolation='nearest')
plt.show()
```

### 通过平均值法进行图像灰度化

```python
A.shape
```

```python
B = A.mean(axis=2)
B = np.tile(B.reshape(B.shape + (1,)), 3).astype(int)
B.shape
```

```python
# 显示灰度化后的图像
plt.imshow(B, cmap=plt.cm.gray, interpolation='nearest')
plt.show()
```

## 使用SVD进行图像重构

```python
# 取前k个特征，对图像进行还原
def get_image_feature(s, k, is_ratio=False):
    num = min(s.shape)
    s_temp = np.zeros(num)
    if is_ratio:  # 如果输入的是比例值
        # 对于S，只保留原特征值数量的一定比例k
        nk = int(k * num)
        s_temp[0: nk] = s[0: nk]
    else:  # 如果输入的是特征个数
        # 对于S，只保留前K个特征值
        s_temp[0: k] = s[0: k]
    s = s_temp * np.identity(num)
    # 用新的s_temp，以及p,q重构A
    temp = np.dot(p, s)
    temp = np.dot(temp, q)
    plt.imshow(temp, cmap=plt.cm.gray, interpolation='nearest')
    plt.show()
```

```python
# 转化成二维矩阵
B = B[:, :, 0]
# 对图像矩阵B进行奇异值分解，得到p,s,q
p,s,q = svd(B, full_matrices=False)
```

### 1. 保留前k个特征值

```python
# 取前k个特征，对图像进行还原
get_image_feature(s, 5)
get_image_feature(s, 50)
get_image_feature(s, 500)
```

### 2. 保留前k比例个特征值

```python
# 取前k个特征，对图像进行还原
get_image_feature(s, 0.01, is_ratio=True)
get_image_feature(s, 0.1, is_ratio=True)
get_image_feature(s, 0.5, is_ratio=True)
```

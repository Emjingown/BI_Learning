```python
import numpy as np
import pandas as pd

from warnings import filterwarnings
filterwarnings('ignore')
pd.set_option('max_columns', None)
```

```python
df_train = pd.read_csv('./data/train.csv')
df_train.sample(3)
```

```python
df_train['Label'] = df_train['Attrition'].map({'No': 0, 'Yes': 1})
df_train.drop('Attrition', axis=1, inplace=True)
df_train.sample(3)
```

```python
df_test = pd.read_csv('./data/test.csv')
df_test['Label'] = -1
df_test.head(3)
```

#### 没有重复值

```python
df_train.duplicated().sum()
```

```python
df_train.describe()
```

```python
df_data = pd.concat([df_train, df_test], axis=0, ignore_index=True)
df_data.sample(5)
```

```python
df_data.nunique()
```

特征列中，user_id、EmployeeNumber都是唯一标识，EmployeeCount、Over18、StandardHours只包含一种结果，因此可去除

```python
cols = [col for col in df_data.columns if col not in ['user_id', 'EmployeeNumber', 'EmployeeCount', 'Over18', 'StandardHours']]
df_data2 = df_data[cols]
df_data2.sample(3)
```

### 数据预处理

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

#### 类别数据编码

```python
df_data2.nunique()[df_data2.nunique()<20]
```

```python
df_data2.sample(3)
```

考虑多次项衍生：Education、EnvironmentSatisfaction、JobInvolvement、JobLevel、JobSatisfaction、PercentSalaryHike、

```python
cat_feas = df_data2.select_dtypes('object').columns
cat_feas
```

```python
df_data2[cat_feas].nunique()
```

类别数据采用独热编码

```python
df_data2 = pd.concat([df_data2, pd.get_dummies(df_data2[cat_feas])], axis=1)
df_data2.drop(columns=cat_feas, inplace=True)
df_data2.sample(5)
```

```python
df_data2.info()
```

#### 数据标准化

```python
train_data = df_data2[df_data2['Label']!=-1]
test_data = df_data2[df_data2['Label']==-1]
y_train = train_data['Label'].values
train_data.drop('Label', axis=1, inplace=True)
test_data.drop('Label', axis=1, inplace=True)
print(train_data.shape, test_data.shape)
```

```python
train_data.head(3)
```

```python
std_sc = StandardScaler()
train_data = std_sc.fit_transform(train_data)
X_test = std_sc.transform(test_data)
```

### 模型训练与评估

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
```

```python
X_tra, X_val, y_tra, y_val = train_test_split(train_data, y_train, random_state=0)
```

```python
lr = LogisticRegression()
lr.fit(X_tra, y_tra)
```

```python
print(f'训练集上的auc为{roc_auc_score(y_tra, lr.predict_proba(X_tra)[:,1]):.4f},验证集上的auc为{roc_auc_score(y_val, lr.predict_proba(X_val)[:,1]):.4f}')
```

### 超参数调整

```python
for c in np.arange(0.1, 2, 0.1):
    lr = LogisticRegression(penalty='l2', C=c)
    lr.fit(X_tra, y_tra)
    print(f'C = {c:.1f}')
    print(f'训练集上的auc为{roc_auc_score(y_tra, lr.predict_proba(X_tra)[:,1]):.4f},验证集上的auc为{roc_auc_score(y_val, lr.predict_proba(X_val)[:,1]):.4f}')
```

```python
for c in np.arange(0.05, 0.46, 0.05):
    lr = LogisticRegression(penalty='l2', C=c)
    lr.fit(X_tra, y_tra)
    print(f'C = {c:.2f}')
    print(f'训练集上的auc为{roc_auc_score(y_tra, lr.predict_proba(X_tra)[:,1]):.4f},验证集上的auc为{roc_auc_score(y_val, lr.predict_proba(X_val)[:,1]):.4f}')
```

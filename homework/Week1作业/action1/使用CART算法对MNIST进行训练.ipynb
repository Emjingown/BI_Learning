```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_digits
pd.set_option('max_columns', None)
```

### 数据读取

```python
raw_data = load_digits()
```

```python
X_data = raw_data.data
y_data = raw_data.target
```

```python
print(X_data.shape, y_data.shape)
```

### 数据探索与预处理

```python
df = pd.DataFrame(np.concatenate((X_data, y_data.reshape((-1,1))), axis=1))
df.info()
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

数据集没有缺失值和重复值，且树模型的训练无需对数据进行归一化处理

### CART决策树模型训练

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score
```

```python
X_tra, X_val, y_tra, y_val = train_test_split(X_data, y_data, train_size=0.7, random_state=0)
print(X_tra.shape, y_tra.shape)
print(X_val.shape, y_val.shape)
```

```python
clf = tree.DecisionTreeClassifier(criterion='gini', max_depth=None, random_state=0)
clf.fit(X_tra, y_tra)
```

```python
y_pred = clf.predict(X_val)
accuracy_score(y_val, y_pred)
```

```python
clf.get_depth()
```

尝试对数据进行标准化

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_sc = StandardScaler()
```

```python
X_tra_sc = std_sc.fit_transform(X_tra)
X_val_sc = std_sc.transform(X_val)
```

```python
clf2 = tree.DecisionTreeClassifier(criterion='gini', max_depth=None, random_state=0)
clf2.fit(X_tra_sc, y_tra)
```

```python
clf2.score(X_val_sc, y_val)
```

```python
clf2.get_depth()
```

从结果来看，标准化后提高效果不明显

#### 尝试处理异常值

```python
def plot_boxplot(data:pd.DataFrame, fea_list=[], ncols=8):
    if isinstance(data, np.ndarray):
        data = pd.DataFrame(data)
    if not fea_list:
        fea_list = data.columns
    df = data[fea_list]
    fea_num = len(fea_list)
    ncols = ncols if fea_num > ncols else fea_num
    nrows = np.ceil(fea_num / ncols)
    figure = plt.figure(figsize=(nrows*3, ncols*4))
    for i in range(fea_num):
        plt.subplot(nrows, ncols, i+1)
        sns.boxplot(df[fea_list[i]], orient='v', width=0.5)
        plt.ylabel(fea_list[i], fontsize=15)
    plt.show()
```

```python
plot_boxplot(X_data)
```

第0、32和39列数据标准差为0，可删除；从图上看，1,6,7,8,15,16,23,24,31,40,47,48,49,55,56,57,63存在较多明显偏离的异常值，可考虑去除。

```python
X_data.shape
```

```python
X_data = np.delete(X_data, [0,1,6,7,8,15,16,23,24,31,32,39,40,47,48,49,55,56,57,63], axis=1)
X_data.shape
```

```python
X_tra, X_val, y_tra, y_val = train_test_split(X_data, y_data, train_size=0.7, random_state=0)
print(X_tra.shape, y_tra.shape)
print(X_val.shape, y_val.shape)
```

```python
clf = tree.DecisionTreeClassifier(criterion='gini', max_depth=None, random_state=0)
clf.fit(X_tra, y_tra)
```

```python
y_pred = clf.predict(X_val)
accuracy_score(y_val, y_pred)
```

```python
clf.get_depth()
```

可见对于CART模型，似乎对异常数据进行简单的删除只会降低模型拟合效果。

#### 调参

```python
for depth in range(3,15):
    clf = tree.DecisionTreeClassifier(criterion='gini', max_depth=depth, random_state=0)
    clf.fit(X_tra, y_tra)
    print(f'最大层数为：{depth}')
    print(f'样本内准确率为{clf.score(X_tra, y_tra):.4f}, 验证集准确率为{clf.score(X_val, y_val):4f}')
```

可见当层数超过8层后出现疑似过拟合的情况

## 数据加载

```python
raw_data = pd.read_csv('./data/SupplyChain.csv', encoding='unicode-escape')
raw_data.head(3)
```

## 数据探索与处理

### 删除冗余特征

```python
# Customer Email、Customer Password和Product Status只有1种取值，去除
raw_data.drop(['Customer Email', 'Customer Password', 'Product Status'], axis=1, inplace=True)
```

```python
# 查看特征相关性
corr_res = raw_data.corr()
corr_res[corr_res == 1]
```

```python
# 完全相同的特征对：Benefit per order和Order Profit Per Order，Order Item Total和Sales per customer，Product Category Id和Category Id，
#                   Order Customer Id和Customer Id，Order Item Cardprod Id和Product Card Id，Order Item Product Price和Product Price
# 可删除其中一个特征
raw_data.drop(['Order Profit Per Order', 'Order Item Total', 'Product Category Id', 'Order Customer Id', 'Order Item Cardprod Id', 
               'Order Item Product Price'], axis=1, inplace=True)
```

### 缺失值情况与处理

```python
raw_data.loc[:, raw_data.isna().any(axis=0)].isna().sum(axis=0) / len(raw_data)
```

```python
# Order Zipcode和Product Description缺失比例太大，删除
raw_data.drop(['Order Zipcode', 'Product Description'], axis=1, inplace=True)
# Customer Lname和Customer Zipcode采用众数填充
raw_data['Customer Lname'].fillna(raw_data['Customer Lname'].mode()[0], inplace=True)
raw_data['Customer Zipcode'].fillna(raw_data['Customer Zipcode'].mode()[0], inplace=True)
```

```python
# 缺失值处理完成
raw_data.isna().any(axis=None)
```

### 聚合处理
因为需预测欺诈订单，因此按订单号进行聚合处理

```python
# 同个订单只有一种取值的字段，取last操作
agg_last = raw_data.columns[(raw_data.groupby('Order Id').nunique()==1).all(axis=0)]
agg_last
```

```python
new_data = raw_data.groupby('Order Id')[agg_last].last().reset_index(drop=True)
new_data.head(3)
```

```python
# 同一订单存在多种取值的字段
print([c for c in raw_data.columns if c not in agg_last])
```

```python
agg_dict = {'Benefit per order': ['sum', 'mean'], 'Sales per customer': ['sum', 'mean'], 'Order Item Discount': ['sum', 'mean'], 
           'Sales': ['sum', 'mean'], 'Order Item Quantity': 'sum', 'Product Price': 'mean', 'Order Item Id': 'count'}
temp = raw_data.groupby('Order Id').agg(agg_dict)
temp.columns = [f'{agg.capitalize()}({k})' for k, agg in temp.columns]
temp.head(3)
```

```python
new_data = new_data.merge(temp, how='left', on='Order Id')
del temp
```

```python
# 计算订单利润率、折扣率
new_data['Order Discount Rate'] = new_data['Sum(Order Item Discount)'] / new_data['Sum(Sales)']
new_data['Order Profit Ratio'] = new_data['Sum(Benefit per order)'] / new_data['Sum(Sales per customer)']
new_data.head(3)
```

### 日期数据处理

```python
date_cols = [c for c in new_data.columns if 'date' in c]
date_cols
```

```python
new_data[date_cols].head(3)
```

```python
for col in date_cols:
    new_data[col] = pd.to_datetime(new_data[col], format='%m/%d/%Y %H:%M')
    new_data[col + '_year'] = new_data[col].dt.year
    new_data[col + '_month'] = new_data[col].dt.month
    new_data[col + '_day'] = new_data[col].dt.day
    new_data[col + '_diff'] = (new_data[col] - new_data[col].min()).dt.seconds / 3600
    new_data.drop(col, axis=1, inplace=True)  # 删除原日期列
new_data.head(3)
```

### 标签编码处理

```python
from sklearn.preprocessing import LabelEncoder
```

```python
cate_cols = new_data.select_dtypes(include='object').columns.tolist()
print(cate_cols)
```

```python
for col in cate_cols:
    if col != 'Order Status':
        lc = LabelEncoder()
        new_data[col] = lc.fit_transform(new_data[col])
new_data.head(3)
```

### 划分训练数据与测试数据
随机抽30%的订单作为测试数据，并对label进行设定

```python
new_data['label'] = (new_data['Order Status'] == 'SUSPECTED_FRAUD').astype('int64')
new_data.drop('Order Status', axis=1, inplace=True)
print(f'总订单数：{new_data["Order Id"].shape[0]}')
print(f'欺诈订单数量：{new_data["label"].sum()}')
```

```python
np.random.seed(2021)
set_order_id = set(new_data['Order Id'])
test_order_id = set(np.random.choice(list(set_order_id), size=int(len(set_order_id)*0.3), replace=False))
train_order_id = set_order_id - test_order_id
```

```python
train = new_data[new_data['Order Id'].isin(train_order_id)].reset_index(drop=True)
test = new_data[new_data['Order Id'].isin(test_order_id)].reset_index(drop=True)
print(len(train), len(test))
```

```python
train_x = train.drop('label', axis=1)
train_y = train['label']
test_x = test.drop('label', axis=1)
test_y = test['label']
```

## 模型训练与预测

```python
import lightgbm as lgb
from sklearn.metrics import roc_auc_score
```

```python
model = lgb.LGBMClassifier(boosting_type="gbdt", num_leaves=30, reg_alpha=0, reg_lambda=0.,
            max_depth=-1, n_estimators=2000, objective='binary',metric= 'auc',
            subsample=0.95, colsample_bytree=0.95, subsample_freq=1,
            learning_rate=0.02, random_state=2021)
```

```python
# model.fit(train_x, train_y, early_stopping_rounds=100, eval_metric='auc', 
#           categorical_feature=cate_feature, sample_weight=train_data.loc[train_idx]['sample_weight'], verbose=5)
model.fit(train_x, train_y, eval_metric='auc')
pred_y = model.predict(test_x)
print('auc =', roc_auc_score(test_y, pred_y))
```

### 可解释性分析

```python
feas_importance = pd.DataFrame(model.feature_importances_, index=model.feature_name_).sort_values(0, ascending=False).T
feas_importance
```

由特征重要性结果来看，最重要的特征与订单所在州、订单城市、客户ID等比较相关，可能的原因是过去有欺诈行为的客户在未来更可能存在欺诈行为，因此与客户ID等高度相关。而像订单利润率、折扣率等也能较好地帮助区分欺诈订单。
